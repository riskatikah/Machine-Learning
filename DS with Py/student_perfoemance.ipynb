```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('unclean_student_performance.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Participation_Score'] = pd.to_numeric(df['Participation_Score'], errors='coerce')
df['Study_Hours_Per_Week'] = pd.to_numeric(df['Study_Hours_Per_Week'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Final_Exam_Score'].fillna(df['Final_Exam_Score'].mean(), inplace= True)
df['Participation_Score'].fillna(df['Participation_Score'].max(), inplace= True)
df['Attendance_Rate'].fillna(df['Attendance_Rate'].max(), inplace= True)
df['Assignment_Score'].fillna(df['Assignment_Score'].mean(), inplace= True)
df['Assignment_Completion_Rate'].fillna(df['Assignment_Completion_Rate'].mean(), inplace= True)
df['Study_Hours_Per_Week'].fillna(df['Study_Hours_Per_Week'].max(), inplace= True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
correlation_matrix = df[['Final_Exam_Score', 'Participation_Score', 'Assignment_Completion_Rate']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Final Exam Score, Participation Score, and Assignment Completion Rate')
plt.show()
```

- Tidak ada korelasi yang kuat antar variabel.
- Participation Score memiliki sedikit hubungan positif dengan Final Exam Score, tetapi tidak terlalu signifikan.
- Assignment Completion Rate memiliki korelasi negatif yang lemah dengan variabel lainnya, menunjukkan bahwa menyelesaikan tugas tidak selalu meningkatkan nilai ujian atau partisipasi kelas.
- Warna lebih merah berarti korelasi lebih tinggi, sementara warna lebih pucat berarti korelasi lebih lemah atau mendekati nol.

```python
# Parental_Education_Level influannce in mean Final_Exam_Score 

plt.figure(figsize=(10, 6))
sns.boxplot(x='Parental_Education_Level', y='Final_Exam_Score', data=df)
plt.title('Final Exam Score by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Final Exam Score')
plt.xticks(rotation=45)  

```

```python
df['Num_Friends'] = df['Num_Friends'].apply(lambda x: x if x <= 20 else 20)

bins = [0, 50, 75, 100] 
labels = ['Low', 'Medium', 'High'] 
df['Final_Exam_Category'] = pd.cut(df['Final_Exam_Score'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Final_Exam_Category', y='Num_Friends', data=df,  palette="coolwarm")
plt.title('Number of Friends by Final Exam Score Category')
plt.xlabel('Final Exam Category')
plt.ylabel('Number of Friends')
plt.show()
```

```python
study_hours_avg_score = df.groupby('Study_Hours_Per_Week')['Final_Exam_Score'].mean().reset_index()

plt.figure(figsize=(20, 6))
plt.bar(study_hours_avg_score['Study_Hours_Per_Week'], study_hours_avg_score['Final_Exam_Score'], color='black')
plt.title('Final Exam Score vs. Study Hours Per Week')
plt.xlabel('Study Hours Per Week')
plt.ylabel('Average Final Exam Score')
plt.xticks(rotation=45)  
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['number'])
corr_matrix = df_numeric.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Matrix")
plt.show()
```

```python
# Buat box plot untuk distribusi skor partisipasi per kelas
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Participation_Score', data=df, palette="coolwarm")

# Judul dan label
plt.xlabel('Class')
plt.ylabel('Participation Score')
plt.title('Distribution of Participation Scores per Class')
plt.xticks(rotation=45)

# Tampilkan plot
plt.show()
```

```python
class_percentages = df.groupby('Class')['Participation_Score'].mean() * 100 / len(df)

plt.figure(figsize=(8, 8))
plt.pie(class_percentages, labels=class_percentages.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Participation Score per Class')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
