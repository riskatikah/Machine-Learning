```python
# count of mechine model in every store
plt.figure(figsize=(12, 6))
sns.countplot(x='Store_ID', hue='Machine_Model', data=df)
plt.title('Number of Machine Models per Store ID')
plt.xlabel('Store ID')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# avg total sales in every store
df_mean = df.groupby(['Coffee_Type', 'Store_ID'])['Total_Sales'].mean().unstack()

plt.figure(figsize=(12, 6))
for store in df_mean.columns:
    plt.plot(df_mean.index, df_mean[store], marker='o', label=f'Store {store}')

plt.title('Average Sales per Coffee Type per Store (Line Plot)')
plt.xlabel('Coffee Type')
plt.ylabel('Average Total Sales')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# store have the high avg sales
df_mean = df.groupby('Store_ID')['Total_Sales'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Store_ID', y='Total_Sales', data=df_mean, marker='o')
plt.title('Average Sales per Store (Line Plot)')
plt.xlabel('Store ID')
plt.ylabel('Average Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Maintenance needed in  every store and what the mechine model

maintenance_by_store_model = df.groupby(['Store_ID', 'Machine_Model'])['Maintenance_Needed'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Store_ID', y='Maintenance_Needed', hue='Machine_Model', data=maintenance_by_store_model)
plt.title('Maintenance Needed per Store and Machine Model')
plt.xlabel('Store ID')
plt.ylabel('Maintenance Needed Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Avg_Daily_Processes and Avg_Error_Per_Day

correlation = df['Avg_Daily_Processes'].corr(df['Avg_Error_Per_Day'])

print(f"Correlation between Avg_Daily_Processes and Avg_Error_Per_Day: {correlation}")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Avg_Daily_Processes', y='Avg_Error_Per_Day', data=df)
plt.title('Relationship between Average Daily Processes and Average Errors per Day')
plt.xlabel('Average Daily Processes')
plt.ylabel('Average Errors per Day')
plt.show()
```

```python
avg_error_by_model_store = df.groupby(['Machine_Model', 'Store_ID'])['Avg_Error_Per_Day'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Machine_Model', y='Avg_Error_Per_Day', hue='Store_ID', data=avg_error_by_model_store)
plt.title('Average Error per Day by Machine Model and Store ID')
plt.xlabel('Machine Model')
plt.ylabel('Average Error per Day')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

heatmap_data = df.pivot_table(values='Avg_Error_Per_Day', index='Machine_Model', columns='Store_ID', aggfunc='mean')

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Heatmap of Average Errors per Day by Machine Model and Store ID')
plt.xlabel('Store ID')
plt.ylabel('Machine Model')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
```

Pada proses indentivay coleration ketika proses harian incress dan error incress ke. jadi sekarang kita harus mengetahui model mesin mana yang sering digunakan di store id dan setelah itu kita harus menambahkan model mesin untuk mengurangi tingkat error yang menyebabkan pengeluaran uang untuk service.
