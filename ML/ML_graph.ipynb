```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
# set figsize
plt.figure(figsize=(4,3))
nx.draw(G)
```

```python
G.add_nodes_from([3,4,5,6])
# add several nodes
plt.figure(figsize=(4,3))
nx.draw(G, with_labels=True)
```

```python
G.add_edge(1,2)
G.add_edge(1,3)
plt.figure(figsize=(4,3))
nx.draw(G, with_labels=True,
        node_color='yellow',
        pos=nx.kamada_kawai_layout(G))
```

```python
edgelist = [(1,4), (1,5), (2,5), (2,6), (3,5)]
G.add_edges_from(edgelist)
plt.figure(figsize=(4,3))
nx.draw(G, with_labels=True,
        node_color='yellow',
        pos=nx.kamada_kawai_layout(G))
```

```python
G.edges()
```

```python
G.remove_edge(3,5)
plt.figure(figsize=(4,3))
nx.draw(G, with_labels=True,
        node_color='yellow',
        pos=nx.kamada_kawai_layout(G))
```

```python
G.remove_node(2)
plt.figure(figsize=(4,3))
nx.draw(G, with_labels=True,
        node_color='green',
        pos=nx.kamada_kawai_layout(G))
```

```python
edges = [(1, 2, 5), (1, 6, 2), (2, 3, 6), (2, 4, 3),
         (2, 6, 5), (3, 4, 6), (3, 5, 1), (4, 8, 9),
         (4, 9, 8), (6, 7, 4)]
G1 = nx.Graph()
G1.add_weighted_edges_from(edges)
plt.figure(figsize=(4,3))
# show the weights

nx.draw(G1, with_labels=True,
        node_color='yellow',
        pos=nx.shell_layout(G1))
```
